```python
import glob

def box_annotation(gun_directory):
    read_dir = glob.glob(gun_directory)
    for read_file in read_dir:
        rename_file = read_file.replace(
            '/home/sparkbrains/Documents/INQ_AI/Models/Yolo_Forklift/data/labels/validation/',
            '')
        flag = 0
        count = 0
        with open(read_file) as file:
            lines = file.readlines()
            lines = [line.rstrip() for line in lines]
            for i in lines:
                #print(i, "----------",read_file)

                li_file = list()
                if i.find("0") == 0:
                    li_file.append("8" + i[1:])
                else:
                    li_file.append("8" + i[1:])

                print(li_file)
                with open(
                        "/home/sparkbrains/Documents/INQ_AI/Models/Yolo_Forklift/data/labels/validation_/" + rename_file,
                        'a') as fd:
                    if flag == 0:
                        fd.write(f'{li_file[0]}')
                        flag = 1
                    else:
                        fd.write(f'\n{li_file[0]}')
                count += 1
        print(count)

gun_directory = '/home/sparkbrains/Documents/INQ_AI/Models/Yolo_Forklift/data/labels/validation/*'

box_annotation(gun_directory)
```

```python
# Importing all necessary libraries
import cv2
import os

# Read the video from specified path
cam = cv2.VideoCapture("/home/cygnet/Documents/Gaurav/DATASET/Models/suspicious_dataset/sleeping_videos/1666871381186508-1673864546-1673865446.mp4")

try:
	
	# creating a folder named data
	if not os.path.exists('/home/cygnet/Documents/Gaurav/DATASET/Models/Yolo_Forklift/frames/'):
		os.makedirs('/home/cygnet/Documents/Gaurav/DATASET/Models/Yolo_Forklift/frames/')

# if not created then raise error
except OSError:
	print ('Error: Creating directory of data')

# frame
currentframe = 0

while(True):
	
	# reading from frame
	ret,frame = cam.read()

	if ret:
		# if video is still left continue creating images
		name = '/home/cygnet/Documents/Gaurav/DATASET/Models/Yolo_Forklift/frames/' + str(currentframe) + '.jpg'
		print ('Creating...' + name)

		# writing the extracted images
		cv2.imwrite(name, frame)

		# increasing counter so that it will
		# show how many frames are created
		currentframe += 1
	else:
		break

# Release all space and windows once done
cam.release()
# cv2.destroyAllWindows()
```
