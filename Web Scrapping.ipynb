```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By

# Create a web driver object
driver = webdriver.Chrome()

# Navigate to the website
driver.get('https://platform.inq.inc/')
```

```python
links = []
images_count = 0
page_counts = 0


    
for j in range(77):
    for i in driver.find_elements(By.XPATH, '//*[@id="tabpanel-3"]/div/div/div[8]/div/table/tbody/tr/td[6]/div/img'):
        links.append(i.get_attribute('src'))
        images_count =+ 1
        time.sleep(2)
    print('link extracted')

    #next page
    driver.find_element(By.XPATH, '//*[@id="tabpanel-3"]/div/div/div[8]/div/div/div[2]/nav/ul/li[9]/button').click()
    page_counts =+ 1
    time.sleep(3)
    print('next page')
    j =+ 1
```

```python
import urllib.request


# Save image
for i,j in enumerate(links):
    urllib.request.urlretrieve(j , f"/home/cygnet/Documents/Gaurav/DATASET/Disagree_images/image_{i}.jpg")
```

```python
import folium

# Create a Map object
m = folium.Map(location=[45.5236, -122.6750], zoom_start=13)

# Create a list of locations
locations = [[45.5236, -122.6750], [45.5236, -122.6623], [45.5434, -122.6623], [45.5434, -122.6750]]

# Create a PolyLine object and add it to the map
folium.PolyLine(locations, color="red", weight=2.5, opacity=1).add_to(m)

# Save the map
m.save('map.html')
```
